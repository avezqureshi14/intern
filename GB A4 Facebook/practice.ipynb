```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("D:\LAB\DSBDA\GB A4 Facebook\dataset_Facebook.csv",sep=';')
df
```

```python
plt.hist(df['Lifetime Post Total Reach'])
plt.xlabel('Lifetime Post Total Reach')
plt.ylabel('Frequency')
plt.show()
```

```python
#boxplot
plt.figure(figsize=(8,6))
sns.boxplot(data=df[['Lifetime Post Total Reach','Lifetime Post Total Impressions','Lifetime Engaged Users']])
plt.title('')
plt.show()
```

```python
#scatterplot
plt.figure(figsize=(15,12))
sns.heatmap(df.corr(),annot=True,cmap='coolwarm',square=True)
plt.title("Scatterplot of Facebook dataset")
plt.show()
```
