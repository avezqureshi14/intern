```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python

df = pd.read_csv(r"D:\LAB\DSBDA\GB A4 Facebook\dataset_Facebook.csv",sep=";")
df
```

Histogram

```python
plt.hist(df['Lifetime Post Total Reach'])
plt.xlabel('Lifetime Post Total Reach')
plt.ylabel('Frequency')
plt.show()
```

Boxplot

```python
sns.boxplot(x='Category', y='Lifetime Post Total Reach', data=df)
plt.show()
```

Scatter plot

```python
plt.scatter(df['Lifetime Post Total Reach'], df['Lifetime Engaged Users'])
plt.xlabel('Lifetime Post Total Reach')
plt.ylabel('Lifetime Engaged Users')
plt.show()
```

Heatmap

```python
pivot_table = df.pivot_table(index='Category', columns='Type', values='Lifetime Post Total Reach', aggfunc='mean')
sns.heatmap(pivot_table, cmap='YlGnBu')
plt.show()
```
