```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Air Quality dataset
air_quality = pd.read_csv('D:\LAB\DSBDA\WD A9 heart and air\AirQualityUCI.csv')

# Load Heart Disease dataset
heart_disease = pd.read_csv('D:\LAB\DSBDA\WD A9 heart and air\heartdisease.csv')
```

FOR AIR QUALITY DATASET

```python
# Plot a histogram of CO(GT) values
plt.hist(air_quality['CO(GT)'], bins=10)
plt.xlabel('CO(GT)')
plt.ylabel('Frequency')
plt.title('Histogram of CO(GT) Values')
plt.show()
```

```python
# Plot a scatter plot of NO2(GT) vs. PT08.S4(NO2)
plt.scatter(air_quality['NO2(GT)'], air_quality['PT08.S4(NO2)'])
plt.xlabel('NO2(GT)')
plt.ylabel('PT08.S4(NO2)')
plt.title('Scatter Plot of NO2(GT) vs. PT08.S4(NO2)')
plt.show()
```

FOR HEART DISEASE DATASET

```python
# Plot a bar chart of the count of each chest pain type
sns.countplot(x='cp', data=heart_disease)
plt.xlabel('Chest Pain Type')
plt.ylabel('Count')
plt.title('Bar Chart of Chest Pain Types')
plt.show()
```
