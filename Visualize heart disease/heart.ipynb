```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('D:\LAB\DSBDA\Visualize heart disease\heartdisease.csv')
df
```

```python
# Histogram
plt.figure(figsize=(8, 6))
plt.hist(df['age'], bins=10, edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Age')
plt.show()
```

```python
# Boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(data=df[['trestbps', 'chol', 'thalach']])
plt.title('Boxplot of trestbps, chol, thalach')
plt.show()
```

```python
# Scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='age', y='chol', hue='num')
plt.title('Scatter Plot of Age vs Cholesterol')
plt.show()
```

```python
# Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Heatmap')
plt.show()
```
