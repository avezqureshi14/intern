```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("D:\LAB\DSBDA\Visualize heart disease\heartdisease.csv")
df
```

```python
#histogram
plt.figure(figsize=(8,6))
plt.hist(df['age'],bins=10,edgecolor='black')
plt.title('Histogram of Age')
plt.show()
```

```python
#Boxplot
plt.figure(figsize=(8,6))
sns.boxplot(data=df[['trestbps','chol','thalach']])
plt.title("Box plot of trestbps,chol, thalach")
plt.show()
```

```python
#scatter plot
plt.figure(figsize=(8,6))
sns.scatterplot(data=df,x='age',y='chol',hue='num')
plt.title('Scatterplot of age vs cholestrol')
plt.show()
```

```python
#heatmap
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(),cmap='coolwarm',annot=True,square=True)
plt.title("Scatter plot of age vs cholestrol")
plt.show()
```
