```python
request1.content
```

```python
soup = bs4.BeautifulSoup(request1.text)
soup
```

Fetching Reviews and Comments

```python
reviews = soup.find_all('div',{'class':'t-ZTKy'});
for review in reviews:
    print(review.get_text()  +  "\n\n")
```

```python
#Average Overall rating
ratings = soup.find('div',{'class':'_3LWZlK'}).get_text();
print(ratings)
```

```python
#Individual Rating
individual_ratings = soup.find_all('div',{'class':'_3LWZlK _1BLPMq'});
for indiv_rating in individual_ratings:
    print(indiv_rating.get_text()  +  "\n\n")
```

Fetching Tags

```python
tags = soup.find('span',{'class':'yhB1nd GXgmTe'}).getText();
tags
```

Fetching Customer Names


```python
customer_name = soup.findAll('span',{'class':'_2GEIK5'});
for cust_name in customer_name:
    print(cust_name.get_text() + "\n")
```
